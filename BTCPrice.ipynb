```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BTCPrice.csv")

# Store the data
df['Date'] = pd.to_datetime(df['Date'])

# Calculate price average
df['Average'] = ((df['High'] + df['Low']) / 2)

print(df['Average'])
```

```python
# It is important to note the Bitcoin trades 24/7, therefore doesn't have "open"
# or "close" prices

# Plot the historical open price
plt.figure(figsize=(15, 5))
plt.plot(df['Date'], df['Average'])
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Open Price')
```
